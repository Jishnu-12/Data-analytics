### Null value handling

```python
c2 = ["usb","hdmi","slim","camera","bluetooth","Wifi","released"]

df["spec_score"] = df["spec_score"].fillna(df["spec_score"].median())

for i in c2:
    df[i] = df[i].fillna(df[i].mode()[0])

df.isnull().sum()
```

### Outlier Removal

```python
c3 = df.select_dtypes(["int","float"])

fig,axs = plt.subplots(1,4,figsize = (20,5))
axs = axs.flatten()

for i in range(4):
    axs[i].boxplot(df[c3.columns[i]])
    axs[i].set_title(c3.columns[i])
plt.show()
```

```python
def outliers(df,col):
    Q1 = np.percentile(df[col],25,method = "midpoint")
    Q3 = np.percentile(df[col],75,method = "midpoint")
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR
    outlier = df[(df[col]>upper)|(df[col]<lower)]
    outlier_list = outlier[col].tolist()
    return upper,lower,outlier,outlier_list
```

```python
upper,lower,outlier,list1 = outliers(df,"price")
print(f"upper={upper}, lower={lower}, list1={list1}")
outlier.head()
```

```python
df.drop(outlier.index,inplace = True)
```

```python
outlier = df[(df["price"]>upper)|(df["price"]<lower)]["price"].tolist()
outlier
```

```python
upper,lower,outlier,list2 = outliers(df,"no_rates")
print(f"upper={upper}, lower={lower}, list2={list2}")
outlier.head()
```

```python
df.drop(outlier.index,inplace = True)
```

```python
outlier = []
for i in df["no_rates"]:
    if((i>upper)or(i<lower)):
        outlier.append(i)
outlier
```

```python
upper,lower,outlier,list3 = outliers(df,"rating_100")
print(f"upper={upper}, lower={lower}, list3={list3}")
outlier.head()
```

```python
df.drop(outlier.index,inplace = True)
```

```python
outlier = df[(df["rating_100"]>upper)|(df["rating_100"]<lower)]["rating_100"].tolist()
outlier
```

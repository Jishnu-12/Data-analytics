```python
for i in ["Price (in rupees)"]:
    df[i] = df[i].fillna(df[i].median())
```

```python
missing = df.select_dtypes("object")
```

```python
for i in missing.columns:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
plot = df["Price (in rupees)"]
plt.boxplot(plot)
plt.show()
```

```python
Q1 = np.percentile(df["Price (in rupees)"],25, method = "midpoint")
Q3 = np.percentile(df["Price (in rupees)"],75, method = "midpoint")
print(f"Q1 = {Q1},Q2 = {Q3}")
```

```python
IQR = Q3-Q1
upper = Q3 + 1.5 * IQR
lower = Q1 - 1.5 * IQR
print(f"upper_limit = {upper} , lower_limit = {lower}")
```

```python
outlier = df[(df["Price (in rupees)"] > upper) | (df["Price (in rupees)"] < lower)]["Price (in rupees)"].tolist()
outlier
```

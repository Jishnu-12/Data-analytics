# Pandas Data Preparation Basics

```python
import pandas as pd
import numpy as np

list1 = {
    'name':['jishnu','vishnu','anusree','swetha','abinav','navadil','ajay'],
    'age':[26,15,25,np.nan,23,np.nan,np.nan],
    'sallery':[25000,np.nan,10000,20000,23000,22000,21000],
}

df = pd.DataFrame(list1) # creating a data frame from a python dictionary
df.fillna({'age':df['age'].mean()},inplace = True) # handling missing values from age column
df.rename(columns = {"sallery":"salary"},inplace = True) # renaming column
df.fillna({'salary':df['salary'].median()}, inplace = True) # handling missing values in salary column
df[['age','salary']] = df[['age','salary']].astype(int) # changing data type of column
df["promotted salary"] = df['salary'] * 2 # adding new column with promotted salary
df["promotted salary"] = df["promotted salary"].astype(int) 
df.index = np.arange(1,8) # coustomizing index lables
df.index.name = "No" # adding name to index lables
df[['salary','promotted salary']] = df[['salary','promotted salary']].astype(object)
df.loc[2,['salary','promotted salary']] = "student" # changing values from a column
print(df)
print()
a = df.loc[:,['age']]
b = df.iloc[:,[0]]
c = df.loc[2].tolist()
df.replace({'ajay':'ajay peter'},inplace = True) # replacing a current value to new value
print(a)
print()
print(b)
print()
print(c)
print()
d = df['age'].value_counts() # used to find out frequency 
print(d)
print()
su =df.loc[:,'age'].sum()
print(f"sum of age column is  = {su.tolist()}")
```

## to_frame()

```python
list1 = [1,2,3,4,5,6,7]
s1 = pd.Series(list1)
s1.name = "numbers"
print(s1)
df = s1.to_frame() # converts a series to DataFrame
print(df)
d = df['numbers'].sort_values(ascending = False).tolist() # sorting in desending order converting it in to list
print(f" desending list {d}")
```

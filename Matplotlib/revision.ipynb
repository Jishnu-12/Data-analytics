```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.get_dataset_names()
df = sns.load_dataset("penguins")

fig,axs = plt.subplots(2,2, figsize =(15,12))

sns.scatterplot(data = df , x="bill_length_mm" ,y ="flipper_length_mm",ax = axs[0,0],hue = "island")
axs[0,0].set_title("scatter plot")
axs[0,0].tick_params(axis = "x",color = "red")
axs[0,0].grid()

axs[0,1].plot(df["bill_length_mm"],color ="black" , marker ="o", mfc = "red")
axs[0,1].set_title("line plot")
axs[0,1].grid(color = "green",linestyle = "--")


columns = ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
sns.heatmap(data = df[columns].corr(),annot = True,ax = axs[1,0],cmap = "coolwarm")
axs[1,0].set_title("heatmap")

d = df.groupby("species")["body_mass_g"].sum()
axs[1,1].pie(d ,labels = d.index,autopct = "%1.0f%%",explode = [0,0.2,0])
axs[1,1].set_title("piechart")

plt.suptitle("combining seaborn and  matplot lib" , fontweight ="bold")
plt.tight_layout()
plt.savefig("combination.jpg",dpi = 250, bbox_inches = "tight")
df.dropna(inplace = True)
```

```python
data = {
    "num1":[3,4,5,2],
    "name":["a","b","c","d"]
}
colors = ["red","lightblue","purple","yellow"]
df= pd.DataFrame(data)
g = df.groupby("name")["num1"].sum()
plt.pie(g,labels= g.index,autopct = "%1.0f%%",colors = colors)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names() #seborn have built in data sets for practicing
```

```python
df = sns.load_dataset("penguins")
```

```python
df["species"].value_counts()
```

```python
df["island"].value_counts()
```

```python
sns.set_style("ticks") #controls the entier figuer
sns.set_context("paper") # controlls the scaling of plots
sns.scatterplot(data = df, x = "flipper_length_mm" , y = "body_mass_g",hue = "island",palette= "Set1")
sns.despine() # removes the spine (border of plot) by default top and right spine will be removed
```

```python
sns.scatterplot(data = df , x = "species" , y = "body_mass_g",hue = "island",style = "sex" , alpha = 0.5)
```

```python
sns.stripplot(data = df , x = "species" , y = "body_mass_g", hue = "island",dodge = True)
```

```python
sns.set_context("paper")
sns.swarmplot(data = df , x = "species" , y = "body_mass_g", hue = "island",palette = "Set1")
sns.despine()
```

```python
sns.histplot(data = df , x = "body_mass_g",hue = "sex",multiple="stack")
```

```python
sns.regplot(data = df, x = "flipper_length_mm", y = "body_mass_g",color = "black")
```

```python
sns.set_style("whitegrid")
sns.lineplot(data = df, x = "body_mass_g", y = "flipper_length_mm",hue ="sex")
```

```python
sns.jointplot(data = df , x = "body_mass_g", y = "flipper_length_mm")
```

```python
sns.set_style("whitegrid")
sns.jointplot(data = df , x = "body_mass_g", y = "flipper_length_mm", kind = "reg",color = "black")
```

```python
sns.barplot(data = df, x = "species" , y = "body_mass_g",hue = "sex" , estimator = np.sum)
```

```python
sns.countplot(data = df, x = "species", hue ="island")
```

```python
sns.boxplot(data = df , x = "species" , y = "body_mass_g", hue = "island")
```

```python
sns.violinplot(data = df , x = "species" , y = "body_mass_g", hue = "sex")
```

```python
sns.violinplot(data = df , x = "species" , y = "body_mass_g", hue = "sex" , split = True)
```

```python
sns.violinplot(data = df , x = "species" , y = "body_mass_g",color = "lightblue")
sns.swarmplot(data = df , x = "species" , y = "body_mass_g", hue = "island", size = 3)
```

```python
sns.kdeplot(data = df, x = "body_mass_g",hue = "species", fill = True)
```

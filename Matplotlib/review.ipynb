Plot a line graph for x = [1,2,3,4,5] and y = [10,20,25,30,40].

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y = [10,20,25,30,40]

plt.plot(x,y,marker = "o",color = "black")
plt.show()
```

Draw a boxplot of tips dataset showing total_bill by day.

```python
import seaborn as sns
import pandas as pd
sns.get_dataset_names()
```

```python
df = sns.load_dataset("tips")

df.head()
```

```python

sns.boxplot(data = df , x = "day" , y = "total_bill")
```

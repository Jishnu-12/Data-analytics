```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num = np.arange(1,11)

plt.plot(num,marker = "o",color = "black",mec = "black",mfc = "red",label = "line")
plt.grid(linestyle = "--",color = "green")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Simple Line Chart 1-10")
plt.legend()
plt.show()
```

```python
x = np.array([2,4,5,6,7,8])

y = x**2

plt.plot(x,y,color = "black",marker = "o" ,mec = "black",mfc = "red")
#plt.plot(y,color = "red",marker= "o",mec = "red",mfc = "black")
plt.show()
```

```python
data  = np.random.normal(1,25,10)
plt.hist(data)
plt.show()
```

```python
x = np.array([2,3,4,5,6])
y = np.array([12,34,52,12,34])
plt.scatter(x,y)
plt.show()
```

```python
x = np.array(["apple","orange","banana","mango","grape","pinapple"])
y = np.array([3,4,5,6,2,1])
plt.bar(x,y,color= ["black","red","black","red","black","red"])
plt.tick_params(axis = "x" , rotation = 90, labelsize = 10)
plt.show()
```

```python
data = {
    "name":["jhon","david","maria"],
    "age":[25,24,22],
    "salary":[15000,25000,18000]
}

df = pd.DataFrame(data)
fig,axs = plt.subplots(1,3, figsize = (15,5))

axs[0].plot(df["name"],df["salary"],color = "black",marker = "o",mfc = "red")
axs[0].set_title("A")
axs[0].set_xlabel("x axis")
axs[0].set_ylabel("y axis")

axs[1].pie(df["age"],labels = df["name"],autopct = "%1.0f%%")
axs[1].set_title("B")
axs[1].legend(fontsize = 6)

axs[2].hist(df["salary"])
axs[2].set_title("C")

plt.tight_layout()

plt.show()

```

```python
data = {
    "name":["jishnu","vishnu","anusree"],
    "age":[26,15,26],
    "salary":[35000,25000,30000]
}

df = pd.DataFrame(data)
fig,axs = plt.subplots(1,3, figsize = (17,5))

#plot1
axs[0].plot(df["name"],df["salary"],color = "black",marker = "o",mec = "black",mfc = "red")
axs[0].set_xlabel("name of employee")
axs[0].set_ylabel("salery of employee")
axs[0].set_title("Name Salary",fontweight = "bold")
axs[0].tick_params(axis = "x",labelsize = 8)
axs[0].grid(color = "green",linestyle ="--")

#plot2
colors = ["red","black","purple"]
axs[1].scatter(df["age"],df["salary"],color = colors)
axs[1].grid(linestyle =":",color = "black")
axs[1].set_xlabel("age",fontsize = 10)
axs[1].set_ylabel("salary")
axs[1].set_title("Age Salary",fontweight = "bold")

#plot3
axs[2].pie(df["salary"],labels= df["name"] , autopct = "%1.0f%%",explode = [0.1,0,0],shadow = True)
axs[2].set_title("Name Salary",y = 1.00)
axs[2].legend(loc = "upper left",fontsize = 8)
axs[2].axis("equal")
fig.suptitle("Sample Subploting",fontweight = "bold",y = 1.05)
plt.tight_layout()

plt.savefig("subploting.jpg",dpi = 250 , bbox_inches = "tight")
```

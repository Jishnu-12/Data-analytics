```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "age":[25,23,26,21,12,34,13,25]
}

df = pd.DataFrame(data)
plt.plot(df["age"], color = "black", marker = "o",markerfacecolor="red", markeredgecolor="black", linestyle = "--",linewidth = "2")
plt.grid()
plt.title("Age chart",color = "red" , fontsize = 16,fontweight = "bold")
plt.show()
```

```python
plt.hist(df["age"])
plt.show()
```

```python
data = [12,23,21,34,54,23,25,26,27,32]
s1 = pd.Series(data)
s1.name = "age"
df = s1.to_frame()
plt.plot(df["age"],color = "black",marker = "o",markerfacecolor = "red",markeredgecolor = "black")
plt.grid(plt.minorticks_on())
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y,marker = "o",color = "black",markerfacecolor = "red",markeredgecolor = "black")
plt.grid()
plt.show()
```

```python
data = {
    "fruits":['apple','orange','pinapple','grape'],
    'sold':[12,24,21,10]
}
df = pd.DataFrame(data)
 

plt.pie(df["sold"], labels = df["fruits"],autopct = "%1.0f%%",explode = [0,0.1,0,0], colors = ['red','orange','yellow','purple'],wedgeprops={"edgecolor":"black", "linewidth":1})
plt.axis("equal")
plt.title("sales report")
plt.show()



plt.bar(df["fruits"], df['sold'], color=['red','darkorange','yellow','purple'])

plt.title("Fruit Sales")
plt.ylabel("Sold Quantity")
plt.show()
```

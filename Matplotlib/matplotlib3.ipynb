```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd


data = {
    "Fruits": ["Apple", "Banana", "Orange", "Mango", "Grapes"],
    "Sold": [50, 75, 30, 100, 65],
    "Price": [120, 40, 60, 150, 90]
}

df = pd.DataFrame(data)
colors = ["red","yellow","orange","green","purple"]
plt.bar(df["Fruits"],df["Sold"],width = 0.6,color = colors,label = ["Apple", "Banana", "Orange", "Mango", "Grapes"])
plt.legend(title = "Price",fontsize = 7,loc = "upper left",labels = df["Price"])
plt.title("Fruits Sales",size = 16,family = "monospace",pad = 25)
plt.xlabel("Fruits",fontweight = "bold")
plt.ylabel("Sales" , fontweight = "bold")
plt.grid( linestyle = "--",color = "black")
plt.grid(which = "minor" , linestyle = ":" , color = "green")
plt.minorticks_on()
plt.show()
```

```python
colors = ["red","yellow","orange","green","purple"]

plt.pie(df["Sold"],labels = df["Fruits"],autopct = "%1.0f%%",shadow = True,explode = [0,0,0,0.2,0],colors = colors)
plt.axis("equal")
plt.title("fruits Sales pie chart",fontweight = "bold",size = 15,pad = 20)
plt.legend(fontsize = 6,labels = df["Price"],title = "price")

plt.show()
```

```python
#plot1
colors = ["red","yellow","orange","green","purple"]
plt.subplot(1,2,1)
plt.bar(df["Fruits"],df["Sold"],width = 0.6,color = colors,label = ["Apple", "Banana", "Orange", "Mango", "Grapes"])
plt.title("A",size = 16,family = "monospace",fontweight = "bold",pad = 10)
plt.xlabel("Fruits",fontweight = "bold")
plt.ylabel("Sales" , fontweight = "bold")
plt.grid( linestyle = "--",color = "black")
plt.tick_params(axis='x', which='major', labelsize=5)
plt.grid(which = "minor" , linestyle = ":" , color = "green")
plt.minorticks_on()
#plot2
colors = ["red","yellow","orange","green","purple"]
plt.subplot(1,2,2)
plt.pie(df["Sold"],labels = df["Fruits"],autopct = "%1.0f%%",shadow = True,explode = [0,0,0,0.2,0],colors = colors,textprops = {"fontsize":7})
plt.axis("equal")
plt.title("B",fontweight = "bold",size = 15,pad = 10)
plt.legend(fontsize = 6,labels = df["Price"],title = "price")
plt.suptitle("Fruits Sales Chart",family = "monospace",size = 16,y = 1.1)
plt.savefig("sample.jpg",bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "age":[25,23,26,21,12,34,13,25]
}

df = pd.DataFrame(data)
plt.plot(df["age"], color = "black", marker = "o",markerfacecolor="red", markeredgecolor="black", linestyle = "--",linewidth = "2")
plt.grid()
plt.title("Age chart",color = "red" , fontsize = 16,fontweight = "bold")
plt.show()
```

```python
plt.hist(df["age"])
plt.show()
```

```python
data = [12,23,21,34,54,23,25,26,27,32]
s1 = pd.Series(data)
s1.name = "age"
df = s1.to_frame()
plt.plot(df["age"],color = "black",marker = "o",markerfacecolor = "red",markeredgecolor = "black")
plt.grid(plt.minorticks_on())
plt.show()
```

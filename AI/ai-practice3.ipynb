```python
review = """""
The product quality is really good and I am satisfied.
However, the delivery was late and the packaging was damaged.
Customer support responded quickly and solved my issue.
"""
```

```python
from nltk import sent_tokenize,word_tokenize
import string
token = sent_tokenize(review)

all_word = []

for i in token:
    word = word_tokenize(i)

    word = [w.lower() for w in word] # lower casting

    word = [w for w in word if w not in string.punctuation] # removes punctuations

    all_word.append(word)

print(all_word)
```

```python
from nltk.corpus import stopwords

stopword = set(stopwords.words('english'))

clean = [
    [w for w in token if w not in stopword]
    for token in all_word
]

print(clean)
```

```python
from nltk.stem import WordNetLemmatizer

lemittizer = WordNetLemmatizer()

lemmatized = [
    [lemittizer.lemmatize(w) for w in sentence]
    for sentence in clean
]

print(lemmatized)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

vectorizer = CountVectorizer()
tfidf = TfidfVectorizer()

documents = [' '.join(sen) for sen in lemmatized]

bow = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(bow.toarray())
```

```python
matrix = tfidf.fit_transform(documents)

print(tfidf.get_feature_names_out())

print(matrix.toarray())
```

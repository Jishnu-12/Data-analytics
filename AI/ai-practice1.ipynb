```python
text = "Hi I am Jishnu. Currently Studying Nlp!. Can I Make it Through?"
```

```python
text = text.lower()

text
```

```python
import string

text = text.translate(str.maketrans('','',string.punctuation))

text
```

```python
from nltk import word_tokenize

token = word_tokenize(text)

token
```

```python
from nltk.corpus import stopwords

stopw = set(stopwords.words('english'))

clean = [w for w in token if w not in stopw]

print(clean)
```

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer

stemmer = PorterStemmer()

lemmitizer = WordNetLemmatizer()
```

```python
text1 = [stemmer.stem(w) for w in clean]
print(text1)
```

```python
text2 = [lemmitizer.lemmatize(w) for w in clean]

text2
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

text3 = ' '.join(text1)

bow = vectorizer.fit_transform([text3])

print(vectorizer.get_feature_names_out())
print(bow.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

text4 = ' '.join(text2)

matrix = tfidf.fit_transform([text4])

print(tfidf.get_feature_names_out())
print(matrix.toarray())
```

```python
text = "I am Jishnu? Loving dataAnalytics ! Completing? playing"
```

```python
from nltk import word_tokenize
import string
from nltk.corpus import stopwords

stopword = set(stopwords.words('english'))
token = word_tokenize(text)
clean1 = [w.lower() for w in token]
clean2 = [w for w in clean1 if w not in string.punctuation]
clean3 = [w for w in clean2 if w not in stopword]

print(clean3)
print("\n")

from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

clean4 = [wnl.lemmatize(w, pos = "v") for w in clean3]

print(clean4)
print("\n")

from sklearn.feature_extraction.text import CountVectorizer

cvr = CountVectorizer()

join1 = [' '.join(clean4)]

bow = cvr.fit_transform(join1)

print(cvr.get_feature_names_out())
print(bow.toarray())
```

```python
from nltk import word_tokenize
import string

text = "hai i Am Jishnu? can i really COMpleate this?"

token = word_tokenize(text)

token
```

```python
repun = [w for w in token if w not in string.punctuation]

repun
```

```python
lower = [w.lower() for w in repun]

lower
```

```python
from nltk.corpus import stopwords

stopword = set(stopwords.words('english'))

clean = [w for w in lower if w not in stopword]
clean
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

text1 = [stemmer.stem(w) for w in clean]
text2 = [lemmatizer.lemmatize(w) for w in clean]

print(text1)
print(text2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer

vectorizer = CountVectorizer()
tfidf = TfidfVectorizer()

document1 = [' '.join(text1)]
document2 = [' '.join(text2)]

bow = vectorizer.fit_transform(document1)
matrix = tfidf.fit_transform(document2)

print(vectorizer.get_feature_names_out())
print(bow.toarray())
print("\n")
print(tfidf.get_feature_names_out())
print(matrix.toarray())
```

```python
text = "Hello Jishnu! I love data analytics, and NLP is interesting."

lower_casting = text.lower() # nlp step 1 lower casting

lower_casting
```

```python
import string

lower_casting = lower_casting.translate(str.maketrans('','', string.punctuation)) # step 2 remove punctuation

print(lower_casting)
```

```python
from nltk import word_tokenize

tokens = word_tokenize(lower_casting) # toknization splitting texts to smaller units

print(tokens)
```

```python
from nltk.corpus import stopwords

stopword = set(stopwords.words('english'))

clean = [w for w in tokens if w not in stopword]

print(clean)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed_words = [stemmer.stem(w) for w in clean]

print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

lemmatize_word = [lemmatizer.lemmatize(w) for w in clean]

print(lemmatize_word)
```

```python
text1 = ' '.join(lemmatize_word)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

bow = vectorizer.fit_transform([text1]) # bag of words

print(vectorizer.get_feature_names_out())
print(bow.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

tfidf_matrix = tfidf.fit_transform([text1])

print(tfidf.get_feature_names_out())

print(tfidf_matrix.toarray())
```

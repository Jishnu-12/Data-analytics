```python
import pandas as pd 

df = pd.read_csv(r"C:\Users\jk977\OneDrive\Desktop\nlp dataset\nlp_reviews_500.csv")
df.head()
```

```python
df.tail()
```

```python
review = df["review_text"]
```

```python
from nltk import word_tokenize
import string

all_word = []

for i in review:
    word = word_tokenize(i)
    word = [w.lower() for w in word]
    word = [w for w in word if w not in string.punctuation]
    all_word.append(word)

all_word
```

```python
from nltk.corpus import stopwords
stopword = set(stopwords.words('english'))

text1 = [[w for w in text if w not in stopword] for text in all_word]
text1
```

```python
from nltk.stem import WordNetLemmatizer,PorterStemmer
wnl = WordNetLemmatizer()
ps = PorterStemmer()

clean1 = [[wnl.lemmatize(w) for w in text] for text in text1] # transforming to dictionary type sqeesing
clean2 = [[ps.stem(w) for w in text] for text in text1] # cutting from roots

print(clean1)
print("\n")
print(clean2) 
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv = CountVectorizer()
tfidf = TfidfVectorizer()

join1 = [' '.join(w) for w in clean1]
join1

bow = cv.fit_transform(join1)

print(cv.get_feature_names_out())
print(bow.toarray())
```

```python
join2 = [' '.join(w) for w in clean1]

matrix = tfidf.fit_transform(join2)

print(tfidf.get_feature_names_out())
print(matrix.toarray())
```

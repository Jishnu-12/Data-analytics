```python
review = """
The product quality is? really good and I am satisfied.
However, the delivery was late and the packaging was !damaged.
Customer support responded quickly and solved my issue.
"""
```

```python
from nltk import sent_tokenize,word_tokenize
import string
from nltk.corpus import stopwords

token = sent_tokenize(review)
stopword = set(stopwords.words('english'))

clean = []
for i in token:
    word = word_tokenize(i)
    word = [w.lower() for w in word]
    word = [w for w in word if w not in string.punctuation]
    word = [w for w in word if w not in stopword]
    clean.append(word)
print(clean)
print("\n")

from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

clean1 = [[wnl.lemmatize(w, pos = "v") for w in sen]for sen in clean]

print(clean1)
print("\n")
from sklearn.feature_extraction.text import TfidfVectorizer

join1 = [" ".join(w) for w in clean1]

tfidf = TfidfVectorizer()

matrix = tfidf.fit_transform(join1)

print(tfidf.get_feature_names_out())
print(matrix.toarray())
```

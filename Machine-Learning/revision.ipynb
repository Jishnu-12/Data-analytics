```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/home/jishnu/Desktop/ML-dataset/ML-practice-dataset.csv")
```

```python
df1 = data.copy()

df1.head()
```

### Linear Regression

```python
x = df1.drop("annual_spending", axis = 1)
y = df1["annual_spending"]

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)

le = LabelEncoder()

x_train["loan_approved"] = le.fit_transform(x_train["loan_approved"])
x_test["loan_approved"] = le.transform(x_test["loan_approved"])

c1 = ["age","income","experience_years","education_years","hours_per_week","credit_score","distance_from_work_km"]

mm = MinMaxScaler()

x_train[c1] = mm.fit_transform(x_train[c1])
x_test[c1] = mm.transform(x_test[c1])

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

model = lr.fit(x_train,y_train)

y_predict = model.predict(x_test)

from sklearn.metrics import r2_score

r2_score(y_test,y_predict)
```

### Logistic regression

```python
df2 = data.copy()
df2.head()
```

```python
x = df2.drop("loan_approved", axis = 1)
y = df2["loan_approved"]

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)

le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

mm = MinMaxScaler()

c2 = ["age","income","experience_years","education_years","hours_per_week","credit_score","distance_from_work_km","annual_spending"]

x_train[c2] = mm.fit_transform(x_train[c2])
x_test[c2] = mm.transform(x_test[c2])

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix

lgr = LogisticRegression()

model = lgr.fit(x_train,y_train)

y_predict = model.predict(x_test)

print(accuracy_score(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
```

### Svm classifier

```python
df3 = data.copy()
df3.head()
```

```python
x = df3.drop("loan_approved",axis = 1)
y = df3["loan_approved"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20, random_state = 42)

le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

mm = MinMaxScaler()

c3 = ["age","income","experience_years","education_years","hours_per_week","credit_score","distance_from_work_km","annual_spending"]

x_train[c3] = mm.fit_transform(x_train[c3])
x_test[c3] = mm.transform(x_test[c3])

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report

svm = SVC(kernel = "linear")

model = svm.fit(x_train,y_train)

y_predict = model.predict(x_test)

print(classification_report(y_test,y_predict))
print(accuracy_score(y_test,y_predict))
```

### KNN (K-Nearest-Neighbours)

```python
df4 = data.copy()
df4.head()
```

```python
x = df4.drop("loan_approved", axis = 1)
y = df4["loan_approved"]

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)

le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

mm = MinMaxScaler()

c4 = ["age","income","experience_years","education_years","hours_per_week","credit_score","distance_from_work_km","annual_spending"]

x_train[c4] = mm.fit_transform(x_train[c4])
x_test[c4] = mm.transform(x_test[c4])

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors = 3)

model = knn.fit(x_train,y_train)

y_predict = model.predict(x_test)

print(accuracy_score(y_test,y_predict))
```

### Decision TRee

```python
df5 = data.copy()
df5.head()
```

```python
x = df5.drop("loan_approved", axis = 1)
y = df5["loan_approved"]

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)

le = LabelEncoder()

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

tr = DecisionTreeClassifier(max_depth = 2)

model = tr.fit(x_train,y_train)

y_predict = model.predict(x_test)

print(accuracy_score(y_test,y_predict))
```

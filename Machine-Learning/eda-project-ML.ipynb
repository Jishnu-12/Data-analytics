```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix
```

```python
df = pd.read_csv("/home/jishnu/Desktop/ML-dataset/eda-projectfile.csv")
```

```python
df.head()
```

```python
c1 = df.select_dtypes("object")
c1.nunique()
```

```python
df.drop(["Name","Platform","Publisher","Developer"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
x = df.drop("Global_Sales" , axis = 1)
y = df[["Global_Sales"]]

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)

c2 = ["Game_category","Age_Rating"]

le = LabelEncoder()

for i in c2:
    x_train[i] = le.fit_transform(x_train[i])
    x_test[i] = le.transform(x_test[i])

mm = MinMaxScaler()

c3 = ["Year_of_Release","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Critic_Score","Critic_Count","User_Score","User_Count"]

x_train[c3] = mm.fit_transform(x_train[c3])
x_test[c3] = mm.transform(x_test[c3])

lr = LinearRegression()

model = lr.fit(x_train,y_train)

y_predict = model.predict(x_test)

r2_score(y_test,y_predict)
```

```python
x = df.drop("Age_Rating",axis = 1)
y = df["Age_Rating"]

x_train,x_test,y_train,y_test = train_test_split(x,y , test_size = 0.20, random_state = 42)

le = LabelEncoder()

x_train["Game_category"] = le.fit_transform(x_train["Game_category"])
x_test["Game_category"] = le.transform(x_test["Game_category"])

y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)

mm = MinMaxScaler()

c1 = ["Year_of_Release","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales","Critic_Score","Critic_Count","User_Score","User_Count"]

x_train[c1] = mm.fit_transform(x_train[c1])
x_test[c1] = mm.transform(x_test[c1])

lgr = LogisticRegression(max_iter = 1000)

model = lgr.fit(x_train,y_train)

y_predict = model.predict(x_test)

print(accuracy_score(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
```

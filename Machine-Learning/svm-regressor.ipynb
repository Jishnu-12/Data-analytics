```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.svm import SVR
```

```python
df = pd.read_csv("/home/jishnu/Desktop/ML-dataset/ML-practice-dataset.csv")
```

```python
df.info()
```

```python
x = df.drop("annual_spending", axis = 1)
y = df[["annual_spending"]]

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20, random_state = 42)

le = LabelEncoder()
mm = MinMaxScaler()
mm1 = MinMaxScaler()
svm = SVR()

x_train["loan_approved"] = le.fit_transform(x_train["loan_approved"])
x_test["loan_approved"] = le.transform(x_test["loan_approved"])

c1 = ["age","income","experience_years","education_years","hours_per_week","credit_score","distance_from_work_km"]

x_train[c1] = mm.fit_transform(x_train[c1])
x_test[c1] = mm.transform(x_test[c1])

y_train = mm1.fit_transform(y_train)
y_test = mm1.transform(y_test)


model = svm.fit(x_train,y_train.ravel())

y_predict = model.predict(x_test)

print(r2_score(y_test,y_predict))

```

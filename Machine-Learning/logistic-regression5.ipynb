```python
import pandas as pd 
```

```python
df = pd.read_csv("/home/jishnu/Desktop/ML-dataset/ML-practice-dataset.csv")
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

```python
le = LabelEncoder()
mm = MinMaxScaler()
```

```python
df["loan_approved"] = le.fit_transform(df["loan_approved"])

c1 = ["annual_spending","distance_from_work_km","credit_score","hours_per_week","education_years","experience_years","income","age"]


df[c1] = mm.fit_transform(df[c1])
```

```python
x = df.drop("loan_approved",axis = 1)
y = df["loan_approved"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.20 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()

model = lgr.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_predict)
```

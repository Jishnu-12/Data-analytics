```python
from sklearn.model_selection import train_test_split

x = df.drop("promotion_status",axis = 1)
y = df["promotion_status"]


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 42)
```

```python
x_train = pd.get_dummies(x_train,dtype = int, drop_first = True)
x_test = pd.get_dummies(x_test,dtype = int,drop_first = True)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import MinMaxScaler 

col = ["age","experience_years","hours_per_week","projects_completed","training_hours","performance_score","monthly_income","annual_bonus"]

mm = MinMaxScaler()

x_train[col] = mm.fit_transform(x_train[col])
x_test[col] = mm.transform(x_test[col])
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()

model = lgr.fit(x_train,y_train)

y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_predict)
```

```python
accuracy_score(y_test,y_predict)
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

```python
data = pd.read_csv("/home/jishnu/Desktop/ML-dataset/ml_practice_dataset_600_rows.csv")

df = data.copy()
```

```python
df.head()
```

```python
df.describe()
```

```python
le = LabelEncoder()

c1 = df.select_dtypes("object")

for i in c1.columns:
    df[i] = le.fit_transform(df[i])
```

```python
df.head()
```

```python
c2 = ["age","experience_years","hours_per_week","projects_completed","monthly_income","performance_score","training_hours","annual_bonus"]

mm = MinMaxScaler()

df[c2] = mm.fit_transform(df[c2])
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop("promotion_status",axis = 1)

y = df["promotion_status"]
```

```python
x_test,x_train,y_test,y_train = train_test_split(x,y, test_size = 0.25, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()


model = log.fit(x_train,y_train)

y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_predict)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_predict)
```

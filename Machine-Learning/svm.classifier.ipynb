```python
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("/home/jishnu/Desktop/ML-dataset/ml_perfect_practice_dataset_700.csv")
```

```python
df.head()
```

```python
x = df.drop("promotion_status", axis = 1)
y = df["promotion_status"]

x_train,x_test,y_train,y_test = train_test_split(x,y ,test_size = 0.20 , random_state = 42)
```

```python
x_test  = pd.get_dummies(x_test,drop_first = True,dtype = int)
x_train = pd.get_dummies(x_train,drop_first = True,dtype = int)
```

```python
x_train

```

```python
c1 = ["age","experience_years","hours_per_week","projects_completed","training_hours","performance_score","monthly_income","annual_bonus"]

mm = MinMaxScaler()

x_train[c1] = mm.fit_transform(x_train[c1])

x_test[c1] = mm.transform(x_test[c1])
```

```python
svm = SVC()

model = svm.fit(x_train,y_train)

y_predict = model.predict(x_test)
```

```python
accuracy_score(y_test,y_predict)
```

```python
confusion_matrix(y_test,y_predict)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predict))
```

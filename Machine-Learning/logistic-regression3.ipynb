```python
import pandas as pd 
```

```python
df = pd.read_csv("/home/jishnu/Desktop/ML-dataset/ml_perfect_practice_dataset_700.csv")
```

```python
df.head()
```

```python
df = pd.get_dummies(df,dtype = int, drop_first = True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

col = ["age","experience_years","hours_per_week","projects_completed","training_hours","performance_score","monthly_income","annual_bonus"]

df[col] = mm.fit_transform(df[col])
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop("promotion_status",axis = 1)

y = df["promotion_status"]


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

lgr = LogisticRegression()

model = lgr.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

confusion_matrix(y_test,y_predict)
```

```python
accuracy_score(y_test,y_predict)
```

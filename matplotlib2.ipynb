# univariate analysis
- focused on one variable at a time
- to understand distribution, frequency or pattern
- fro numerical datas we use histogram,boxplot,lineplot,
- for categorical data we use piechart and bar chart

## univariate Numerical

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Salary": [25000, 32000, 28000, 45000, 50000,
               38000, 42000, 36000, 30000, 47000,
               39000, 41000, 35000, 34000, 48000,
               36000, 40000, 43000, 37000, 44000]
}

df = pd.DataFrame(data)

plt.plot(df["Salary"],color = "black",marker = "o",mec ="black",mfc = "red",ms = 10)
plt.title("Univariate Numerical Line Chart")
plt.grid()
plt.show()
```

```python
plt.hist(df["Salary"])
plt.title("Univariate Numerical Histogram")
plt.show()
```

```python
plt.boxplot(df["Salary"])
plt.title("Univariate Numerical Box Chart")
plt.show()
```

## univariate categorical

```python
df["Dept"] = ["HR", "IT", "Finance", "Sales", "Marketing",
              "IT", "HR", "HR", "IT", "IT",
              "IT", "Finance", "HR", "HR", "Marketing",
              "IT", "Finance", "HR", "HR", "HR"]
count = df["Dept"].value_counts()
plt.pie(count,labels = count.index,autopct = "%1.0f%%",shadow = True,explode = [0.1,0,0,0,0],startangle = 90)
plt.axis("equal") # used for making the piechart complete circle
plt.title("Univariate Categorical Pie Chart")
plt.legend(loc = "upper left" , fontsize = 7)
plt.savefig("first pie chart.jpg") # used for saving files 
plt.show()
```

```python
plt.bar(count.index,count,width = 0.5,color = ["black","red","black","red","black"])
plt.title("Univariate Count PLot / Bar Chart")
plt.show()
```

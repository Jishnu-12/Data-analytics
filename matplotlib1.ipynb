```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
x = np.array([0,6,8,10,12])
y  = np.array([0,25,23,54,22])

font1 ={"family":"monospace","color":"black","size":15}
style1 = {"color":"black","marker":"o","ms":12,"mec":"black","mfc":"black"}
style2 ={"color":"red","marker":"o","ms":10,"mec":"black","mfc":"red"}
# ms = markersize
# mec = markeredge colour
# mfc = marker phase colour
plt.plot(x,y,marker = "o",color = "black",ms=15,mec = "black",mfc = "w")
plt.xlabel("Id",fontdict = font1,loc= "right")
plt.ylabel("Age",fontdict = font1,loc = "bottom")
plt.title("Id/Age - Chart",fontdict = font1,loc = "left")
plt.show()
```

```python
plt.plot(x,y,"o-.", ms = 7, mec = "black", mfc = "black" ,color ="black")
plt.show()
```

```python
plt.plot(x,y,linestyle = "dashed",**style1) # unpacking dictionary
plt.show()
```

```python
x1 = np.array([3,6,1,10])
y1 = np.array([6,2,7,11])

plt.plot(x1,**style1)
plt.plot(y1,**style2)
plt.xlabel("Id",fontdict = font1, loc = "right")
plt.ylabel("Age",fontdict = font1 , loc = "bottom")
plt.title("Id V/S Age",fontdict = font1 , loc = "left" , fontweight = "bold")
plt.grid(color = "green",linestyle = "--")
plt.show()
```

```python
x = np.array([1,2,3,4,5,6,7])
y = np.array([21,12,34,25,35,23,10])

colors = ["red","black","purple","green","blue","orange","grey"]
sizes = [20,50,35,45,25,39,100]
# s = size
plt.scatter(x,y , color = colors, s = sizes)
plt.title("Scatter chart",fontweight = "bold" , loc = "left")
plt.show()
```

```python
x = ['a','b','c','d']

y = [2,4,6,8]
colors = ['black','red','black','red']
plt.bar(x,y,color = colors)
plt.title("Bar Chart",fontweight = "bold",loc = "left")
plt.show()
```

```python
x = ['a','b','c','d','e']

y = [10,12,14,16,18]
colors = ['red','black','red','black','red']
plt.barh(x,y,color = colors) # used to create horizontal bar graphs
plt.title("Horizontal Bar Chart",fontweight = "bold",loc = "left")
plt.show()
```

```python
x = ['a','b','c','d']
y = [2,4,6,8]

colors = ['white','red','white','red']
plt.subplot(1,2,1)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("A")
plt.bar(x,y,color = colors,width = 0.5)


x1= ['a','b','c','d','e']
y1 = [10,12,14,16,18]

colors = ['red','white','red','white','red']
plt.subplot(1,2,2)
plt.barh(x1,y1,color = colors , height = 0.5)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("B")
plt.suptitle("Bar chart subploting")
plt.style.use("dark_background") # fo changing the background color
plt.show()

```

```python
x = [10,12,34,25]

label = ['a','b','c','d']
mc = ["red","purple","black","orange"]
plt.pie(x,labels = label,explode = [0,0,0.1,0],startangle = 90,shadow = True,colors = mc)
plt.legend(title = "alphabets",fontsize = 6,loc = "upper left")
plt.show()
```
